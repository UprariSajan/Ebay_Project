```python
import pandas as pd
import requests                     
from bs4 import BeautifulSoup
```

```python
webpage=requests.get('https://www.ebay.com/sch/i.html?_from=R40&_nkw=oppo+phone&_sacat=0&LH_TitleDesc=0&_pgn=2')
#webpage=requests.get('https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2499334.m570.l1313&_nkw=phones&_sacat=9355')
```

```python
webpage.content
```

```python
soup=BeautifulSoup(webpage.content,'html.parser')
```

```python
divs=soup.find_all('div',class_='s-item__info clearfix')
```

```python
len(divs)
```

```python
lst1=[]    
for i in soup.find_all('div',class_='s-item__title'):
    lst1.append(i.text.strip()) 
```

```python
lst2=[]
for i in soup.find_all('span',class_='s-item__price'):
    lst2.append(i.text.strip())
```

```python
lst3=[]
for i in soup.find_all('span',class_='s-item__shipping s-item__logisticsCost'):
    lst3.append(i.text.strip())
```

```python
lst4=[]
for i in soup.find_all('span',class_='s-item__location s-item__itemLocation'):
    lst4.append(i.text.strip())
```

```python
lst5=[]
for i in soup.find_all('span',class_='BOLD'):
    lst5.append(i.text.strip()) 
```

```python
product_name=[]
price=[]
shipment_charges=[]
manufacturing=[]
unit_sold=[]
        
for i in divs: 
            product_name.append(i.find('div',class_='s-item__title').text.strip())  
            price.append(i.find('span',class_='s-item__price').text.strip())
            shipment_charges.append(i.find('span',class_='s-item__shipping s-item__logisticsCost'))
            manufacturing.append(i.find('span',class_='s-item__location s-item__itemLocation'))
            unit_sold.append(i.find('span',class_='BOLD'))

d={'product_name':product_name,'price':price,'shipment_charges':shipment_charges,'manufacturing':manufacturing,'unit_sold':unit_sold}
df=pd.DataFrame(d)
df
```

```python
df=pd.DataFrame(list(zip(lst1,lst2, lst3,lst4,lst5)),
              columns=['product_name','price','shipment_charges','manufcturing','unit_sold'])
```
